```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pyplot_themes as themes
```

# Data Taken for Analysis

Contains : 

- IceCube-59__Search_for_point_sources_using_muon_events
- AMANDA_7_Year_Data
- IceCube-40_String_Data
- first_year_of_IC86_upgoing_events

```python
data_59_string = pd.read_csv('IceCube-59__Search_for_point_sources_using_muon_events.csv')
data_86_string_upgoing = pd.read_csv('first_year_of_IC86_events.csv')
```

```python
RA_data_59_string = data_59_string['RA(deg)']
RA_data_86_string_upgoing = data_86_string_upgoing['RA']
Dec_data_59_string = data_59_string['Dec.(deg)']
Dec_data_86_string_upgoing = data_86_string_upgoing['Dec']

themes.theme_solarized(scheme="light")
#themes.theme_ggplot2(figsize=[10, 5])
#themes.theme_solarized_light2()

f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex = 'col', sharey = 'row')

ax1.boxplot(RA_data_59_string)
ax2.boxplot(RA_data_86_string_upgoing)
ax3.boxplot(Dec_data_59_string)
ax4.boxplot(Dec_data_86_string_upgoing)

f.text(0.28, 0.02, '59 String Data', ha='center', va='center')
f.text(0.03, 0.70, 'Right Ascension (deg)', ha='center', va='center', rotation='vertical')
f.text(0.75, 0.02, '86 String Data', ha='center', va='center')
f.text(0.03, 0.28, 'Declination Angle (deg)', ha='center', va='center', rotation='vertical')
```

```python
x_59_string = data_59_string['RA(deg)']
y_59_string = data_59_string['Dec.(deg)']

plt.hist2d(x_59_string, y_59_string, bins = 120)
plt.colorbar()
plt.show()
```

```python
x_86_string = data_86_string_upgoing['RA']
y_86_string = data_86_string_upgoing['Dec']

themes.theme_solarized(scheme="dark")

plt.hist2d(x_86_string, y_86_string, bins = 120)
plt.colorbar()
plt.show()
```

```python
x_59_string = data_59_string['RA(deg)']
y_59_string = data_59_string['Dec.(deg)']
x_86_string = data_86_string_upgoing['RA']
y_86_string = data_86_string_upgoing['Dec']

themes.theme_solarized(scheme="dark")

f, ((ax1, ax2)) = plt.subplots(1, 2, sharex = 'col', sharey = 'row')

ax1.hist2d(x_59_string, y_59_string, bins = 120)
ax2.hist2d(x_86_string, y_86_string, bins = 120)
```

```python
x_86_string = data_86_string_upgoing['RA']
y_86_string = data_86_string_upgoing['log10(E)']

themes.theme_solarized(scheme="dark")

plt.hist2d(x_86_string, y_86_string, bins = 120)
plt.colorbar()
plt.show()
```

```python
x_86_string = data_86_string_upgoing['log10(E)']
y_86_string = data_86_string_upgoing['Dec']

themes.theme_solarized(scheme="dark")

plt.hist2d(x_86_string, y_86_string, bins = 120)
plt.colorbar()
plt.show()
```

```python
x_59_string = data_59_string['RA(deg)']
y_59_string = data_59_string['log10(MuE)']

plt.hist2d(x_59_string, y_59_string, bins = 120)
plt.colorbar()
plt.show()
```

```python
x_59_string = data_59_string['log10(MuE)']
y_59_string = data_59_string['Dec.(deg)']

plt.hist2d(x_59_string, y_59_string, bins = 120)
plt.colorbar()
plt.show()
```

```python
RA_data_59_string = data_59_string['RA(deg)']
RA_data_86_string_upgoing = data_86_string_upgoing['RA']
Dec_data_59_string = data_59_string['Dec.(deg)']
Dec_data_86_string_upgoing = data_86_string_upgoing['Dec']
E_data_59_string = data_59_string['log10(MuE)']
E_86_string = data_86_string_upgoing['log10(E)']

themes.theme_solarized(scheme="dark")

f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharey = 'row')

ax1.hist2d(RA_data_59_string, E_data_59_string, bins = 70)
ax2.hist2d(RA_data_86_string_upgoing, E_86_string, bins = 70)
ax3.hist2d(Dec_data_59_string, E_data_59_string, bins = 70)
ax4.hist2d(Dec_data_86_string_upgoing, E_86_string, bins = 70)

f.text(0.38, 0.9, 'log10 of Energy Deposit vs Angle')
f.text(0.28, 0.02, '59 String Data', ha='center', va='center')
f.text(0.06, 0.70, 'Right Ascension (deg)', ha='center', va='center', rotation='vertical')
f.text(0.75, 0.02, '86 String Data', ha='center', va='center')
f.text(0.06, 0.28, 'Declination Angle (deg)', ha='center', va='center', rotation='vertical')
```
