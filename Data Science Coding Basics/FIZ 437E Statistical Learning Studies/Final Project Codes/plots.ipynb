```python
import pandas as pd
import numpy as np
from matplotlib.colors import LogNorm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_59_string = pd.read_csv('IceCube-59__Search_for_point_sources_using_muon_events.csv')
data_59_string.head()
```

```python
x = data_59_string['RA(deg)']
y = data_59_string['Dec.(deg)']

heatmap, xedges, yedges = np.histogram2d(x, y, bins=50)
extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]

plt.clf()
plt.imshow(heatmap.T, extent=extent, origin='lower')
plt.title('IceCube-59 Search for point sources data heatmap')
plt.xlabel('Right Ascencion Angle in degrees')
plt.ylabel('Declination Angle in degrees')
plt.show()
```

heatmap histogram, bar eksik!

```python
x = data_59_string['RA(deg)']
y = data_59_string['Dec.(deg)']

plt.hist2d(x, y, bins = 40, norm = LogNorm())
plt.colorbar()
plt.show()
```

hocanın kodu ile plotlanmış hali

```python
x = data_59_string['RA(deg)']
y = data_59_string['Dec.(deg)']

plt.hist2d(x, y, bins = 100)
plt.colorbar()
plt.show()
```
