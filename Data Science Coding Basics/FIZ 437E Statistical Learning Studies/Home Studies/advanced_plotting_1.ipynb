####Plotting Stacked Bar Charts

```python
A = [5., 30., 45., 22.]
B = [5., 25., 50., 20.]
X = range(4)

plt.bar(X, A, color = "b")
plt.bar(X, B, color = "r", bottom = A)
plt.show()
```

```python
A = np.array([5.,30.,45.,22.])
B = np.array([5.,25.,50.,20.])
C = np.array([4.,5.,5.,5.])
x = np.arange(4)

plt.bar(x, A, color = 'r')
plt.bar(x, B, color = 'b', bottom=A)
plt.bar(x, C, color = 'g', bottom=A+B)
plt.show()
```

#### Plotting back-to-back bar charts

```python
women_pop = np.array([5., 30., 45., 22.])
men_pop = np.array([5., 25., 50., 20])
X = np.arange(4)

plt.barh(X, women_pop, color = 'r')
plt.barh(X, -men_pop, color = 'b')
plt.show()
```

####Color Maps (cmap)

```python
import matplotlib.cm as cm

N = 256
angle = np.linspace(0, 8 * 2 * np.pi, N)
radius = np.linspace(.5, 1., N)
X = radius * np.cos(angle)
Y = radius * np.sin(angle)

plt.scatter(X, Y, c = angle, cmap = cm.hsv)
plt.show()
```

# Controlling a line pattern 

```python
N = 8

A = np.random.random(N)
B = np.random.random(N)
X = np.arange(N)

plt.bar(X, A, color = 'r', edgecolor = 'k', hatch = 'x')
plt.bar(X, A + B, bottom = A, edgecolor = 'k', color = 'b', hatch = '/')
plt.show()
```

# Controlling a Marker's Properties

```python
X = np.linspace(-6, 6, 1024)
Y = np.sinc(X)

plt.plot(X, Y,
        linewidth = 3.,
        color = 'k',
        markersize = 9,
        markeredgewidth = 1.5,
        markerfacecolor = '.75',
        markeredgecolor = 'k',
        marker = 'o',
        markevery = 32)
plt.show()
```

```python
X = np.linspace(-6, 6, 1024)
Y = np.sinc(X)

plt.plot(X, Y,
        linewidth = 3.,
        color = 'r',
        markersize = 3,
        markeredgewidth = 1.,
        markerfacecolor = '.50',
        markeredgecolor = 'b',
        marker = '^',
        markevery = 50)
plt.show()
```

```python
from cycler import cycler
import matplotlib

matplotlib.rcdefaults()

x = np.linspace(0, 2 * np.pi)
offsets = np.linspace(0, 2 * np.pi, 4, endpoint = False)

yy = np.transpose([np.sin(x + phi) for phi in offsets])

plt.rc('lines', linewidth = 4)
plt.rc('axes', prop_cycle = (cycler('color', ['r', 'g', 'b', 'y']) + 
                            cycler('linestyle', ['-', '--', ':', '-.'])))
fig, (ax0, ax1) = plt.subplots(nrows = 2)
ax0.plot(yy)
ax0.set_title('Set default color cycle to rgby')

ax1.set_prop_cycle(cycler('color', ['c', 'm', 'y', 'k']) + 
                    cycler('lw', [1, 2, 3, 4]))
ax1.plot(yy)
ax1.set_title('Set axes color cycle to cmyk')

fig.subplots_adjust(hspace = 0.3)
plt.show()
```
