# Private variables and methods

```
class Airplane:
  fuel = 0
  maxFuel = 24000

  def addFuel(self, volume):
    self.fuel = self.fuel + volume
  
  def printStatus(self):
    print('Current fuel: ', self.fuel)
```

```
airbus = Airplane()
airbus.addFuel(20000)
airbus.printStatus()
```

## Adding fuel with "dot"
We have changed value of fuel with a method as you have notices. But there is a way to change fuel value without using a method, as below.

```
airbus.fuel = 22000
airbus.printStatus()
```

However, there is a problem as we have exceeded the maximum fuel volume allowed. We can solve this problem by prohibiting the direct access to the fuel attribute. In the airplane class, add two underscores in front of the fuel attribute.

```
class Airplane:
  __fuel = 0
  maxFuel = 24000

  def addFuel(self, volume):
    self.__fuel = self.__fuel + volume
  
  def printStatus(self):
    print('Current fuel: ', self.__fuel)
```

Adding those two underscores makes the attribute private. 

```
airbus = Airplane()
airbus.printStatus()
airbus.__fuel = 22000
airbus.printStatus()
```

In here we try to change fuel attribute directly again. But notice how it did not change since now this attribute is private.
